```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv(r"C:\Users\Luis Arbio\Documents\Hoteles\data\Processed\reservas_hoteles_processed.csv")
```

```python
df.head(3)
```

```python
# --- Convertir target a 0/1 ---
df['booking_status'] = df['booking_status'].map({'Not_Canceled':0, 'Canceled':1})
```

```python
df.describe()
```

```python
df['booking_status'].value_counts(normalize=True)
```

```python
# --- Ingeniería de atributos ---
df['total_guests'] = df['no_of_adults'] + df['no_of_children']
df['stay_duration'] = df['no_of_weekend_nights'] + df['no_of_week_nights']
df['is_long_stay'] = (df['stay_duration'] > 7).astype(int)
df['special_requests_flag'] = (df['no_of_special_requests'] > 0).astype(int)
```

```python
# --- Variables categóricas a One-Hot Encoding ---
categorical_cols = ['type_of_meal_plan', 'room_type_reserved', 'market_segment_type']
df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)
```
